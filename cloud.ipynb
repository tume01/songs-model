```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/pre_train.csv', names=['lyric', 'tag'])
df.head()
```

```python
text = ""
for index, row in df.iterrows():
    if row[1] == 'breakup':
        text += row[0]
len(text)
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import matplotlib.pyplot as plt
```

```python
stopwords = set(STOPWORDS)
break_mask = np.array(Image.open("./breakup2.png"))
break_mask
```

```python

wc = WordCloud(background_color="white", max_words=3000, mask=break_mask,
               stopwords=stopwords)
wc.generate(text)

# store to file
wc.to_file("cloud.png")

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(break_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
